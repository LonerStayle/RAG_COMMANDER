```python
import os, sys
from pathlib import Path

src_path = Path(os.getcwd()).resolve().parents[1]  
sys.path.append(str(src_path))
```

```python
from langchain_community.document_loaders import Docx2txtLoader
from tools.rag.chunker.default_chunker import chunk_with_recursive 
from utils.util import get_project_root
import subprocess

path = get_project_root() / "src" / "data" / "housing_pre_promise"/ "주택공급에 관한 규칙(국토교통부령)(제01504호)(20250630).docx"

def load_doc(path: str):
    p = Path(path)
    if p.suffix == ".doc":
        # LibreOffice로 doc → docx 변환
        subprocess.run(["libreoffice", "--headless", "--convert-to", "docx", str(p)], check=True)
        p = p.with_suffix(".docx")
    return Docx2txtLoader(str(p)).load()

docs = load_doc(path)
documents = chunk_with_recursive(docs)
```
