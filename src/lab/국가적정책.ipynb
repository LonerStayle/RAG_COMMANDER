```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
국토교통부 부동산 정책 보도자료 크롤러 (2024-2025)

사용법:
    python real_estate_scraper.py
    
필요 라이브러리:
    pip install requests beautifulsoup4 pandas openpyxl
"""

import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
import re

# 기본 설정
BASE_URL = "https://www.korea.kr/briefing/pressReleaseList.do"
HEADERS = {'User-Agent': 'Mozilla/5.0'}

def is_real_estate_policy(title):
    """부동산 관련 국가 정책인지 판단"""
    
    # 제외할 키워드
    exclude_keywords = ['동정', '[차관', '[장관', '리콜']
    if any(kw in title for kw in exclude_keywords):
        return False
    
    # 부동산 핵심 키워드
    real_estate_keywords = [
        # 주택
        '주택', '아파트', '공동주택', '단독주택', '빌라',
        
        # 공급/분양
        '분양', '공급', '청약', '입주', '공공주택', '임대주택',
        '매입임대', '전세임대', '행복주택', '영구임대',
        
        # 개발/건설
        '신도시', '택지', '주택단지', '주거단지', '주거지역',
        '재개발', '재건축', '정비사업', '도시정비',
        
        # 가격/시장
        '부동산', '주택가격', '분양가', '전세', '월세', '임대료',
        '공시가격', '실거래가', '시세',
        
        # 제도/정책
        '주택정책', '부동산정책', '공급대책', '주거복지',
        '청년주택', '신혼부부', '생애최초',
        
        # 금융
        '주택담보대출', '전세자금', '주택구입', 'LTV', 'DTI'
    ]
    
    return any(kw in title for kw in real_estate_keywords)

def get_page(page_num, year):
    """특정 연도 페이지 가져오기"""
    params = {
        'repCode': 'A00006',  # 국토교통부
        'pageIndex': page_num,
        'startDate': f'{year}-01-01',
        'endDate': f'{year}-12-31'
    }
    
    try:
        response = requests.get(BASE_URL, params=params, headers=HEADERS, timeout=10)
        return response.text
    except Exception as e:
        print(f"오류: {e}")
        return None

def parse_page(html):
    """페이지 파싱"""
    soup = BeautifulSoup(html, 'html.parser')
    items = []
    
    # 링크 찾기
    links = soup.select('a[href*="pressReleaseView"]')
    
    for link in links:
        title = link.get_text(strip=True)
        url = "https://www.korea.kr" + link['href'] if link['href'].startswith('/') else link['href']
        
        # 날짜 찾기
        parent = link.parent.parent if link.parent else link
        date_text = parent.get_text()
        
        # 날짜 추출
        date_match = re.search(r'(\d{4})-(\d{2})-(\d{2})', date_text)
        if date_match:
            date = date_match.group(0)
            
            # 부동산 정책만 필터링
            if is_real_estate_policy(title):
                items.append({
                    '날짜': date,
                    '제목': title,
                    '링크': url
                })
    
    return items

def crawl_year(year, max_pages=50):
    """특정 연도 크롤링"""
    print(f"\n{'='*80}")
    print(f"{year}년 데이터 수집 중...")
    print(f"{'='*80}")
    
    year_data = []
    empty_count = 0  # 연속으로 빈 페이지 카운트
    
    for page in range(1, max_pages + 1):
        print(f"  페이지 {page:2d} ...", end=' ')
        
        html = get_page(page, year)
        if not html:
            print("실패")
            break
        
        items = parse_page(html)
        
        if items:
            year_data.extend(items)
            empty_count = 0
            print(f"✓ 부동산 정책 {len(items)}건")
        else:
            empty_count += 1
            print("데이터 없음")
            
            # 연속 3페이지 데이터 없으면 종료
            if empty_count >= 3:
                print(f"  → {year}년 데이터 수집 완료")
                break
        
        time.sleep(0.5)  # 서버 부하 방지
    
    print(f"\n{year}년 총 {len(year_data)}건 수집")
    return year_data

def main():
    print("=" * 80)
    print("국토교통부 부동산 정책 보도자료 크롤러 (2024-2025)")
    print("=" * 80)
    print("수집 대상:")
    print("  - 주택/아파트 관련 정책")
    print("  - 분양/공급/청약 관련")
    print("  - 신도시/재개발/재건축")
    print("  - 부동산 가격/시장 정책")
    print("  - 주거복지/임대주택")
    print("-" * 80)
    
    all_data = []
    
    # 2024년 크롤링
    data_2024 = crawl_year(2024, max_pages=50)
    all_data.extend(data_2024)
    
    # 2025년 크롤링
    data_2025 = crawl_year(2025, max_pages=50)
    all_data.extend(data_2025)
    
    # 결과 저장
    if all_data:
        df = pd.DataFrame(all_data)
        
        # 연도 컬럼 추가
        df['연도'] = df['날짜'].str[:4]
        
        # 최신순 정렬
        df = df.sort_values('날짜', ascending=False)
        
        # 컬럼 순서
        df = df[['날짜', '연도', '제목', '링크']]
        
        # 엑셀 저장
        filename = '국토교통부_부동산정책_2024_2025.xlsx'
        
        with pd.ExcelWriter(filename, engine='openpyxl') as writer:
            # 전체
            df.to_excel(writer, sheet_name='전체', index=False)
            
            # 연도별
            df_2024 = df[df['연도'] == '2024']
            df_2025 = df[df['연도'] == '2025']
            
            if len(df_2024) > 0:
                df_2024.to_excel(writer, sheet_name='2024년', index=False)
            if len(df_2025) > 0:
                df_2025.to_excel(writer, sheet_name='2025년', index=False)
        
        print("\n" + "=" * 80)
        print("✓ 크롤링 완료!")
        print("=" * 80)
        print(f"파일명: {filename}")
        print(f"총 수집: {len(df)}건")
        print(f"  - 2024년: {len(df_2024)}건")
        print(f"  - 2025년: {len(df_2025)}건")
        print("=" * 80)
        
        # 샘플 출력
        print("\n부동산 정책 보도자료 샘플 (최신 10건):")
        print("-" * 80)
        for idx, row in df.head(10).iterrows():
            print(f"[{row['날짜']}] {row['제목']}")
        
    else:
        print("\n데이터가 수집되지 않았습니다.")

if __name__ == "__main__":
    main()
```
